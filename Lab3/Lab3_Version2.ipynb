```python
% matplotlib inline
from math import exp
from matplotlib.pylab import plt
import numpy as np

# Declare variables and constants
lamb = 1
N0 = 100
deltat = 0.01
time = 0
t0 = 0
n = 100
nuclei = [N0]
exact = [N0]
t_axis = [t0]

# Define the functians so it can be called them later to calculate decay over time
def num_of_nuclei(Ni,lam,t):
    """Accept arbitrary numerical arguments and return the exact number of nuclei at time t.
    The arguments are the initial number, lambda, and the time you want to measure at. The 
    result after performing the calculations below is the exact number of nuclei at the time 
    input."""
    
    N = Ni*exp(-lam*t) 
    
    return N

def approx_num_nuclei(Ni,lam,dt):
    """Accept arbitrary numerical arguments and return the approximate number of nuclei at time t.
    The arguments are the initial number, lambda, and the time you want to measure at. The result 
    after performing the calculations below is the approximate number of nuclei at the time input."""
        
    N = Ni + -lam*Ni*(dt)
    
    return N

# Call  the functions above to take 1000 points for both the approximate and actual number functions two create two arrays 
# that can be plotted together to test how accurate the approximation is.
n = N0
for i in range(1000):
    n = approx_num_nuclei(n,lamb,deltat)
    n1 = num_of_nuclei(N0,lamb,time)
    if(n>=0):
        nuclei.append(n)
        exact.append(n1)
        t_axis.append(time)
        time += deltat
    

# Plot the two functions and add labels, a legend, x limits, and a title
plt.plot(t_axis,nuclei,".",markevery = 20,label = "Numerical Approximation")
plt.plot(t_axis,exact,"g-",label = "Actual Decay")
plt.xlabel("Time (s)")
plt.ylabel("Number of Nuclei")
plt.title("Approximate vs Actual Decay")
plt.legend(loc = 'best')
plt.xlim(0,10)
plt.show()
```

The graph above is a graph that displays the decay of a nuclei over time. The blue dots are the points that we created using the numerical approximation of the derivative. We did so by running a loop that used the function to return the number of nuclei based on the change in time and the previous Number of nuclei. The graph above evaluates 1000 points until there is no nuclei left. While in the loop it also calculates the actual decay using the function that creates an exponential decay. The green line above it the curve that resembles the actual decay that is obtained from the exponential function where the blue dots is a marker that is located every 20 points which shows the numerical approximation. When comparing the graphs it seems that the approximation is very accurate when running 100 loops.

```python
# Declare Variables 
lamb = 2
N0 = 100
deltat = 0.01
time = 0
t0 = 0
n = 100
nuclei = [N0]
exact = [N0]
t_axis = [t0]
lamb1 = 1
lamb2 = 1
N0B = 0
deltat1 = 0.01
time1 = 0
t01 = 0
n2 = 100
nuclei1 = [N0B]
t_axis1 = [t01]

# Define the functians so it can be called them later to calculate decay over time
def approx_two_nuclei(NiA,NiB,lamA,lamB,dt):
    """Accept arbitrary numerical arguments and return the approximate number of nuclei at time t.
    The arguments are the initial number for nuclei A and B, lambda for A and B, and the time you 
    want to measure at. The result after performing the calculations below is the approximate number 
    of nuclei from B at the time input."""
        
    N = NiB + lamA*NiA*(dt) - lamB*NiB*(dt)
    
    return N

# Call  the functions above to take 1000 points for the decay of B or until there are no nuclei left, and to place these values 
# into an array along with creating an array for the time at each step.
for i in range(1000):
    n = approx_num_nuclei(n,lamb,deltat)
    n1 = num_of_nuclei(N0,lamb,time)
    if(n>=0):
        nuclei.append(n)
        exact.append(n1)
        t_axis.append(time)
        time += deltat


n2 = 0
n3 = N0B
for i in range(1000):
    n3 = approx_two_nuclei(nuclei[n2],n3,lamb1,lamb2,deltat1)
    if(n2>=0):
        nuclei1.append(n3)
        t_axis1.append(time1)
        time1 += deltat1
        n2 += 1
        n = N0
        
        
# Plot the two decays, A and B, and add labels, a legend, x limits, and a title.
plt.plot(t_axis1,nuclei1,"b-",label = "Decay of B")
plt.plot(t_axis,exact,"g-",label = "Decay of A")
plt.xlabel("Time (s)")
plt.ylabel("Number of Nuclei")
plt.title("Double Nuclei Decay vs Time")
plt.legend(loc = 'best')
plt.xlim(0,10)
plt.show()
```

The graph above shows the decay of A and B. The green line is A and it is the exact number function used to make the green line in the first graph. This is an exponential graph where A decays exponentially and independently of B. B on the otherhand is the blue line. It starts with 0 nuclei but is dependent on A as seen in the function above that returns the number of nuclei at each time using the numerical approximation. The graph shows that as A decays it adds the nuclei lost to B until A and B are equal. Then when equal they both begin to decay exponentially.
